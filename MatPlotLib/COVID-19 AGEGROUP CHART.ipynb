```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid_data=pd.read_csv('AgeGroupDetails.csv')
covid_data.drop(9,inplace=True)
covid_data
```

```python
AgeGroup=covid_data['AgeGroup']
TotalCases=covid_data['TotalCases']
Percentage=covid_data['Percentage']
```

```python
fig1=plt.figure()
ax1=fig1.add_axes([0,0,1,1])
ax1.pie(TotalCases,  labels=AgeGroup, autopct='%1.1f%%', startangle=90,rotatelabels=True)

ax1.axis('equal') 

plt.show()
```

```python


ax = plt.subplot(111, projection='polar')
bars = ax.bar(TotalCases, AgeGroup)

# Use custom colors and opacity
for r, bar in zip(TotalCases, bars):
    bar.set_facecolor(plt.cm.viridis(r / 10.))
    bar.set_alpha(0.5)

plt.show()
```
